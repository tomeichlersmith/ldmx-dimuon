```python
from pathlib import Path
import pickle
import matplotlib as mpl
import mplhep
mpl.style.use(mplhep.style.ROOT)
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('results/calib/all-hits/hist.pkl','rb') as f:
    o = pickle.load(f)
h = o['hist']
cell_id = np.array(h.axes[0])
cell_layer = (cell_id >> 17) & 0x3f
cell_x, cell_y = o['to_cell_center'](cell_id)
```

```python
keep = cell_layer == 6
plt.hexbin(
    cell_y[keep],
    cell_x[keep],
    C = h[:,'muon'].values()[keep],
)
```
